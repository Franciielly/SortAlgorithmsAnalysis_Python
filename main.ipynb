### Geração de números aleatórios

```python
import random

constant_100     = 10**2
constant_1000    = 10**3
constant_10000   = 10**4
constant_100000  = 10**5
constant_1000000 = 10**6

def generate_random_numbers(quantity):
    random_numbers = []
    for _ in range(quantity):
        random_numbers.append(random.randint(1, 100))
    return random_numbers
```

## Merge Sort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    left_index = right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged
```

### Execução

```python
values = generate_random_numbers(constant_1000)
sorted_values = merge_sort(values)

print("Desordenado:", values)
print("Ordenado:", sorted_values)
```
